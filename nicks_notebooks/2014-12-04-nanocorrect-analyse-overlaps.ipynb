Aim: 
    
* Assess the impact of multiple rounds of correction in terms of percentage identity and alignability
* Understand whether there is sufficient overlap information in the dataset to produce contiguous assemblies
* To see whether there are any edge effects or low-quality 'banding' effects

Plan:
    
* Re-run correction on new R7.3 v1.9 data with 5 iterations
* Align each to reference
* Visualise in IGV and look at align fraction length and identity on each round of correction


Start first correction round:

```python
!make -f pipeline.make INPUT=FC20.wf1.9.2D.pass.fasta NAME=FC20
!python makerange.py FC20.pp.fasta > ranges.txt
!samtools faidx FC20.pp.fasta
!cat ranges.txt | parallel python nanocorrect.py FC20 > FC20_corrected.fasta
```
