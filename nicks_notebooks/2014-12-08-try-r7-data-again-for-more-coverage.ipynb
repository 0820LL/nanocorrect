Currently the assemblies are topping out at around N50 ~100kb. Not bad, but not where we need to be.

Although nanocorrect could perhaps be improved by playing with DALIGN settings, lack of coverage may well be a problem. PacBio suggest 20x is the basic minimum for doing this stage (50-100x required for Quiver correction more than overlapping). We have around 10x.

Two possible ways forward:
*   Change DALIGN settings to find more overlaps
*   Add R7 coverage -- this wasn't very successful previously but perhaps better now that nanocorrect has been improved?

The problem with R7 data is that it is generally much worse quality. One compromise might be to use the R7 data to find overlaps but not as 'seed' reads, in order to rescue more of the R7.3 data.

Ultimately we need to generate some more full 2D data from R7.3 for E. coli. We might be able to do that as early as Friday.
